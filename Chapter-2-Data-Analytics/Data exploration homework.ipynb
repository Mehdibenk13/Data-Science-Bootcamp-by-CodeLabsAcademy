# Data Exploration with Pandas
<br>

### Importing the video games sales data set

```python
import pandas as pd
 
df = pd.read_csv("vgsales.csv")
df.head()
```

<br><br><br>


### We drop the sales columns that aren't from EU, so that we can focus on that region only



```python
df.drop(["NA_Sales","JP_Sales","Other_Sales","Global_Sales"], inplace=True, axis=1)
df.head(10)
```

<br><br><br>

## Calculating statistical quantities that describe sales in Europe :


```python
print("Mean = ",df["EU_Sales"].mean())

print("\nMedian = ",df["EU_Sales"].median())

print("\nMode (Most sold game) = \n\n",df.iloc[df["EU_Sales"].mode()])

print("\nVariance = ", df["EU_Sales"].var())

print("\nStandard deviation = ", df["EU_Sales"].std())
```

<br><br><br><br>

# Plotting to visualize distribution of sales in EU
We could use visualization libraries such as <i> matplotlib</i>, but pandas works just fine with its built-in plotting possibilities

## Scatter plot : 
A scatter plot can be used in several ways to highlight different aspects of our data. <br>
In our video games sales set, a scatter plot could reprensent the sales volume in Europe depending on the release year of the game and the name of the game, or maybe depending on the year and the genre of the game.<br> <br>
Here we have chosen to represent the sales volume depending on <b>the year of release</b> and the <b>platform</b> or console that runs the game, <br>
We can extract conclusions from our scatter plot on the evolution of consoles usage throughout the years.

```python
df.plot.scatter(x='Year',
                y='Platform',
                s=df["EU_Sales"]*20, 
                figsize=(10,8),
                title='Game sales volume in Europe')
```

<br><br>
## Histogram
Representation of the sales volume distribution :

```python
df["EU_Sales"].hist(bins=1000,color="blue",figsize=(20,5))
```

<br><br>
# Box plot 
A box plot lets us visualize the percentiles in a box form, where the box is formed by 1st and 3rd quartile and has a separation of the median (2nd quartile) in between <br>
It also lets us visualize outliners a circular form, the outliners are defined as elements that are (1.5 times the interquartile range (the box length) ) far from the box 

```python
df.boxplot(column="EU_Sales",figsize=(5,200))
```

Remark that we made the plot very long so that we can actually see the box of quartiles out of the outliner values
